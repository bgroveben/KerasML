```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

## 1. Load Data

```python
# Fix random seed for reproducibility.
seed = 7
numpy.random.seed(seed)
```

```python
# Load Pima Indians Diabetes dataset from UCI Machine Learning repository.
# http://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
dataset
```

```python
# Split dataset into input (X) and output (Y) variables.
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
X
```

```python
Y
```

## 2. Define Model

```python
# Create Model.
# The first, input, layer has 12 neurons and expects 8 input variables.
# The second, hidden, layer has 8 neurons.
# The third, output, layer has 1 neuron to predict the class (onset of diabetes or not).
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

## 3. Compile Model

```python
# Compile model.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4. Fit Model

```python
# Fit the model.
# This is the part where your CPU/GPU gets a workout.
model.fit(X, Y, nb_epoch=150, batch_size=10)
```

## 5. Evaluate Model

```python
# Evaluate the model.
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

## 6. Tie It All Together

```python
# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, nb_epoch=150, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

## 7. Bonus: Make Predictions

```python
# This is a complete example that makes predictions for each record in the training data.

# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, nb_epoch=150, batch_size=10,  verbose=2)
# calculate predictions
predictions = model.predict(X)
# round predictions
rounded = [numpy.round(x[0]) for x in predictions]
print(rounded)
```
